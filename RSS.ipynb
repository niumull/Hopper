# ZiMuZu

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from datetime import timedelta, date
import time
```

```python
web_url = 'http://www.zimuzu.tv/html/top/week_fav_list.html'
page = requests.get(web_url)
soup = BeautifulSoup(page.content, 'html.parser')
a = soup.find_all('strong')
```

```python
c = soup.find_all('a', target="_blank")
list_top = []
urls_top = []
for item in c:
    if item.text != '':
        list_top.append(item.text)
        urls_top.append('http://www.zimuzu.tv' + item['href'])
list_top = list_top[2:-3]
urls_top = urls_top[2:-3]
most_saved_list = pd.DataFrame()
most_saved_list['Names'] = list_top
most_saved_list['Sites'] = urls_top
most_saved_list = most_saved_list.style.set_properties(**{'text-align': 'left'})
most_saved_list
```
